```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
all_data = pd.read_csv('all_data.csv')
print(all_data.head())
```

```python
print(all_data.Country.unique())
```

```python
print(all_data.Year.unique())
```

```python
all_data = all_data.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
```

```python
print(all_data.head())
```

```python
plt.figure(figsize=(8,6))
sns.displot(all_data.GDP, rug = True, kde=False)
plt.xlabel("GDP in Trillions of U.S. Dollars")
```

```python
plt.figure(figsize=(8,6))
sns.displot(all_data.LEABY, rug = True, kde=False)
plt.xlabel("Life expectancy at birth (years)")
```

```python
dfMeans = all_data.drop("Year", axis = 1).groupby("Country").mean().reset_index()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="LEABY", y="Country", data=dfMeans)
plt.xlabel("Life expectancy at birth (years)")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="GDP", y="Country", data=dfMeans)
plt.xlabel("GDP in Trillions of U.S. Dollars")
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=all_data.Year, y=all_data.GDP, hue=all_data.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("GDP in Trillions of U.S. Dollars")
```

```python
graph = sns.FacetGrid(all_data, col="Country", col_wrap=3,
                      hue = "Country", sharey = False, sharex = False)
graph = (graph.map(sns.scatterplot,"LEABY", "GDP")
         .add_legend()
         .set_axis_labels("Life expectancy at birth (years)", "GDP in Trillions of U.S. Dollars"))
```
